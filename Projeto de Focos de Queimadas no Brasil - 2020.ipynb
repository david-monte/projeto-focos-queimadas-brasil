```python
import pandas as pd
import numpy as np
import plotly.express as px
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('Datasets/Focos_2019-01-01_2019-12-31.csv', usecols=['datahora','estado','municipio','bioma','diasemchuva','precipitacao','riscofogo','latitude','longitude','frp'])
#df = pd.read_csv('C:Datasets/Focos_2019-01-01_2019-12-31.csv', usecols=['datahora','estado','municipio','bioma','diasemchuva','precipitacao','riscofogo','latitude','longitude','frp'])
df
```

```python
df.dropna(subset=['riscofogo'], inplace=True)
```

```python
df[df['frp'].isna()]
```

```python
indexNames = df[(df['riscofogo'] == -999)].index
df.drop(indexNames, inplace=True)
```

```python
df[(df['diasemchuva'] == -999)]
```

```python
df.describe()
```

```python
medianaDiaSemChuva = df[df['diasemchuva'] != -999]['diasemchuva'].median()
```

```python
df['diasemchuva'] = df['diasemchuva'].replace(-999,medianaDiaSemChuva)
```

```python
df.info()
```

```python
df['datahora'] = pd.to_datetime(df['datahora'], format="%Y/%m/%d %H:%M:%S")
```

```python
medianFRP = df[~df['frp'].isna()]['frp'].median()
```

```python
df['frp'] = df['frp'].replace(np.nan,medianFRP)
```

```python
df.info()
```

```python
df_agrupado = df.groupby(['datahora','estado','municipio']).agg({    
    'bioma':'first',
    'diasemchuva':np.median,
    'precipitacao':np.median,
    'riscofogo':np.median,
    'latitude':'first',
    'longitude':'first',
    'frp':np.median
}).reset_index()
df_agrupado.head(50)
```

```python
df_agrupado.info()
```

```python
df_agrupado.describe()
```

```python
df_agrupado
```

```python
# NormalizaÃ§Ã£o por Min-Max no intervalo [1,10]
df_agrupado['frp_norm'] = round((((df_agrupado['frp'] - df_agrupado['frp'].min())/(df_agrupado['frp'].max()-df_agrupado['frp'].min())) * 9) + 1, 2)
```
